## Результаты Анализа:

1. Общая Тенденция Рынка Труда:
   - Я заметил, что данная профессия является не сильно востребованной за границами Москвы. Превосходящее количество вакансий находится именно в москве, не менее 320 из 500. Исходя из этого можно сделать выаводы, что либо профессия востребованна не везде, либо она является новой.

2. Требования к Кандидатам:
   - Проанлизировав требования к кандидатам разного уровня подготовки можно выделить несколько самых основных требования: знание программирования(ЯП, компьютерная граммостность), знание математики(мат.статистика). Для более квалифицированных работников требования повышаются, нужно повышать свои знания в математике, разных ЯП(python,R, SQL), также у более опытных сотрудников должны быть хорошо развиты soft-скиллы(умение общаться и т.д)

3. Тенденции и Прогнозы:
   - Тенденции замеченные мной - низкий порог входа, потому что требования к сотрудникам не слишком высоки, а вакансий много. Прогнозы, которые я бы хотел предположить, это высокое развитие данной профессии, потому что бизнес без вырочки будет бессмыслен, а если нет готовой бизнес модели и плана по развитию, то соответсвенно нет и выручки.

## Выводы и Рекомендации:

- Выводы, которые можно сделать после проделанного анализа - рост вакансий по BI-аналитике стримительно растет. С этим ростом растут и требования для кнадидатов на вакансию. Чем больше зарплата у вакансии, тем болье должен уметь сотрудник, изучая все более и более профессиональные интсрументы: работу с разными програмамми, умение писать рабочий и качественный код.
- Рекомендации, которые я бы хотел дать кандидатам, любой кандидат на информационную профессию должен прокачивать свои hard навыки ежедневно, изучая что-то новое. IT-рынок растет слишком быстро, вытесняя не квалифицированных рабочих, поэтому каждый день нужно открывать для себя что-то новое и улучшать свои старые навыки.

## Ограничения Анализа:

- Ограничения возникшее при анализе возникли с количеством предоставленных данных, их было слишком мало для качественного ответа. Мало того, что было мало данных так еще и присуствуют пропущенные значения в имеющихся данных, что усложнило работу, напрмиер с такими характеристиками - заработная плата, многие работодатели просто не указывают этот пункт в вакансии, а из-за этого становитс яенвозможным качество давать анализ заработной платы по каким-то запросам.
