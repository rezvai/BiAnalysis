# 1. Предобработка данных для построение графа

```python
# Создаем пустой список для хранения данных
jobs_skills = []

# Открываем файл для чтения
with open('dataModuleB.txt', 'r') as f:
    # Читаем строки файла по очереди
    for line in f:
        # Разбиваем строку на массив строк
        items = line.strip()[1:-1].split(', ')
        # Добавляем массив строк в список
        jobs_skills.append(items)
```

```python
# Из библиотеки импортиурем модуль для подсчета пар
from collections import defaultdict

# Словарь для хранения количества встречаемости пар навыков вместе
pair_counts = defaultdict(int)

# Подсчет частоты встречаемости пар навыков
for job in jobs_skills:
    for i in range(len(job)):
        for j in range(i+1, len(job)):
            # Пары сортируются для учета перестановок
            pair = tuple(sorted([job[i], job[j]])) 
            pair_counts[pair] += 1

# Счетчик для ограничения вывода
count = 0
# Вывод результатов
for pair, count in pair_counts.items():
    # Пример вывода
    print(f"Навыки {pair[0]} и {pair[1]} упоминаются вместе {count} раз")
    count += 1
    if count == 5:
        break
```

```python
# Импортиурем библиотеку для работы с данными
import pandas as pd
# Импортиурем библиотеку для работы над строками
import re

# Ваш список данных
data = pair_counts
# Разделение ключей
keys = []
# Разделение значений
values = []

# Заполняем созданные ранее массивы
for key, value in data.items():
    keys.append(key)
    values.append(value)

# Подготовка данных для DataFrame
df_data = {'Столбец 1': [re.sub("[\"']", '', key[0]) for key in keys],
           'Столбец 2': [re.sub("[\"']", '', key[1]) for key in keys],
           'Столбец 3': values}

# Создаем DataFrame
df = pd.DataFrame(df_data)

# Сохраняем DataFrame в файл Excel
df.to_excel('data.xlsx', index=False)
```
